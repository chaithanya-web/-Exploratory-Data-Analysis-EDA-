```python
# Histogram for each numerical column
df.hist(bins=30, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include = ['int64','float64']).corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
df_sample = df.sample(n=2000, random_state=1)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_sample, x='median_income', y='median_house_value', alpha=0.5)
plt.title('Median Income vs Median House Value (Sampled)')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Boxplot: Median house value across different ocean proximity (if available)
if 'ocean_proximity' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='ocean_proximity', y='median_house_value', data=df)
    plt.title('House Value by Ocean Proximity')
    plt.xlabel('Ocean Proximity')
    plt.ylabel('Median House Value')
    plt.xticks(rotation=45)
    plt.show()
```
